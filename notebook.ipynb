```python
# import libraries
import pandas as pd
import requests
```

```python
#import csv file to environ
def data():
    data=pd.read_csv("all_countries.csv")
    return data
```

```python
# import api data to environ
def data1():
    demo = requests.get("https://api.worldbank.org/v2/country/?format=json", headers={"Accept": "application/json"})
    response = demo.json()
    data1=pd.json_normalize(response[1])
    return data1
```

```python
data1()
```

```python
print(data)
```

```python
# Clean data csv file
def cleaned_data(data=data()):
    data = data.fillna("Unknown")
    data = data.drop_duplicates()
    data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_')
    data['area'] = pd.to_numeric(data['area'], errors='coerce')
    data['population'] = pd.to_numeric(data['population'], errors='coerce')
    return data
```

```python
cleaned_data()
```

```python
#Clean data1 api data
def cleaned_data1(data1=data1()):
    data1 = data1.drop_duplicates()
    data1.columns = data1.columns.str.strip().str.lower().str.replace(' ', '_')
    data1['latitude'] = pd.to_numeric(data1['latitude'], errors='coerce')
    data1['longitude'] = pd.to_numeric(data1['longitude'], errors='coerce')
    data1 = data1.rename(columns={'name': 'country', 'capitalcity': 'capital'})
    data1 = data1.drop(columns=[
    'adminregion.id', 'adminregion.iso2code', 'adminregion.value',
    'region.id', 'region.iso2code',
    'incomelevel.id', 'incomelevel.iso2code',
    'lendingtype.id', 'lendingtype.iso2code'
])
    return data1
```

```python
cleaned_data1()
```

```python
#Merge data and data 1 together
def merged():
    data = cleaned_data()
    data1 = cleaned_data1()
    merged = pd.merge(data, data1, on='country', how='inner')
    return merged
```

```python
merged()
```

```python
#convert merged data to a csv
merged = merged()
merged.to_csv('merged_data.csv')
```
